```python
from dbfread import DBF
import pandas as pd

dbf_file_path = 'tbl_big_ntee_birds_point.dbf'

# Open the DBF file
dbf_table = DBF(class_path+dbf_file_path, encoding='utf-8')

# Iterate through records and print them
print("Records from DBF file:")
for record in dbf_table.records:
    print(record)

# Optional: Convert to a Pandas DataFrame for easier data manipulation
df = pd.DataFrame(dbf_table.records)
print("\nDataFrame from DBF file:")
print(df.head())
```
